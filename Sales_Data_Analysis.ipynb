Exploratory Data Analysis

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data=df, hue= 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum()
```

```python
sales_gen =df.groupby(['Gender'], as_index=False)['Amount'].sum()
ax = sns.barplot(x='Gender', y='Amount' , data = sales_gen, hue= 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(x = 'Age Group', data=df, hue= 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age =df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
ax = sns.barplot(x='Age Group', y='Amount' , data = sales_age, hue='Age Group')
```

```python
sales_age =df.groupby(['Age Group'], as_index=False)['Amount'].count().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(10,5)})
ax = sns.barplot(x='Age Group', y='Amount', data = sales_age, hue='Age Group')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state =df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,4)})
ax = sns.barplot(x='State', y='Orders', data = sales_state, hue = 'State')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state =df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,4)})
ax = sns.barplot(x='State', y='Amount', data = sales_state, hue = 'State')
```

```python
ax = sns.countplot(x = 'Marital_Status', data=df, hue= 'Gender')
sns.set(rc={'figure.figsize':(4,2)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df_sorted = df.sort_values(by='Occupation', ascending=False)
ax = sns.countplot(x='Occupation', data=df_sorted, hue='Occupation')
sns.set(rc={'figure.figsize':(22,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
occupation_counts = df['Occupation'].value_counts()
sorted_occupations = occupation_counts.index

ax = sns.countplot(x='Occupation', data=df, order=sorted_occupations, hue='Occupation')
sns.set(rc={'figure.figsize':(22,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
product_counts = df['Product_Category'].value_counts()
sorted_products = product_counts.head(10).index

ax = sns.countplot(x='Product_Category', data=df, order=sorted_products, hue='Product_Category')
sns.set(rc={'figure.figsize':(22,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```
